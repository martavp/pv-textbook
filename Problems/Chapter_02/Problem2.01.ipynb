We will use the package [matplotlib.pyplot](https://matplotlib.org/stable/index.html) to plot the results.


```python
import matplotlib.pyplot as plt
import numpy as np
```

We start by defining the location, date and time. 
We will implement the calculation for every hour on May 12, 2022. 

```python
h = 6.63*10**(-34) # [JÂ·s] Planck constant
e = 1.60*10**(-19) #C electron charge
c = 299792458 #[m/s] Light speed
k_B = 1.38*10**(-23)  # Boltzmann constant
T = 5760 #K Temperature
wls = 10**(-9)*np.arange(1,3000) # wavelengths 
M = [10**(-9)*(2*h*c**2/wl**5)*1/(np.exp(h*c/(k_B*wl*T))-1) for wl in wls]
```

```python
plt.plot(10**9*wls, M, label='5780 K')
plt.ylabel('Spectral radiance ($Wm^{-2}nm^{-1}$)')
plt.xlabel('wavelength (nm)')
plt.xlim(0,3000)
plt.legend()
```
